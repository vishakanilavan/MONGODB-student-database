## Creating  New convenient collections for querying:

```python
x=mycol.aggregate([
             {"$project":{
                 "_id":1,
                 "name":1,
                 "examscore":{"$arrayElemAt":["$scores.score",0]},
                 "queryscore":{"$arrayElemAt":["$scores.score",1]},
                 "homeworkscore":{"$arrayElemAt":["$scores.score",2]}
             }}
])
a=[]
for i in x:
    a.append(i)
print(a)
```

```python

mycol2=db["students1"]
mycol2.insert_many(a)
```

```python
for x in mycol2.find():
    print(x)
```

## Average of all scores in all categories:

```python
#1)Average of exam marks
avg1=mycol2.aggregate([
    {"$group":{"_id":"null",
     "avgexamscore":{"$avg":"$examscore"}}},
])
for i in avg1:
    print(i["avgexamscore"])
```

Taking avgexammark and rounding off as 50

```python
#2)Avg quiz marks
avg2=mycol.aggregate([     #anpther method for cal avg of quizfield with another collection "mycol" 
    {"$unwind":"$scores"},
    {"$match":{"scores.type":"quiz"}},
    {"$group":{"_id":"null",
     "avgquizscore":{"$avg":"$scores.score"}}},
])
for i in avg2:
    print(i["avgquizscore"]) 
```

Taking avgquizmark and rounding off as 50

```python
#3)Avg homework marks:
avg3=mycol2.aggregate([
    {"$group":{"_id":"null",
     "avghomeworkscore":{"$avg":"$homeworkscore"}}},
])
for i in avg3:
    print(i["avghomeworkscore"])
```

Taking avghomeworkmark and rounding off as 70

### 1) Student who scored max in all scores(Exam,Quiz,Homework):  

```python
MaxTotalscore=mycol.aggregate([    
    {"$project":{
    "_id":1,
     "name":1,
     "Totalscores":{"$sum":"$scores.score"}}},
    {"$sort":{"Totalscores":-1}},
    {"$limit":1}
    
])
```

```python

for i in MaxTotalscore:
    print (i["name"])


```

### 2) Students who scored below average in the exam:

```python
query2=mycol.aggregate([
    {"$unwind":"$scores"},
    {"$match":{"$and":[{"scores.type":"exam"},{"scores.score":{"$lt":49}}]}}  #49 is avgexamscore
])
```

```python
for x in query2:
    print(x["name"])
```

### 3)Find students who scored below pass mark and assigned them as fail, and above pass mark as pass in all the categories.

```python
query3=mycol2.aggregate([
    {"$project":{     
        "name":1,
         "Result":{
             "$cond":{
                 "if":{"$and":[{"$gte":["$queryscore",40]},{"$gte":["$examscore",40]},{"$gte":["$homeworkscore",40]}]},
                 "then":"pass",
                 "else":"fail"
    
    }}}}
])
```

```python
for i in query3:
    print(i)
```

### 4) Find the total and average of the exam, quiz and homework and store them in a separate collection.

```python
query4=mycol.aggregate([
    {"$unwind":"$scores"},
    {"$group":{
        "_id":{"id":"$_id","name":"$name"},
        "Total":{"$sum":"$scores.score"},
        "Average":{"$avg":"$scores.score"}
    }},
     {"$project":{
         "_id":0,
         "Id":"$_id.id",
         "Name":"$_id.name",
         "Totalscore":"$Total",
        "Averagescore":"$Average"
    }},
    {"$sort":{"Id":1}}
])
```

```python
a=[]
for i in query4:
    print(i)
    a.append(i)

```

```python
print(a)
```

```python
mycol4=db["TotalandAverage"] #creating new collection for storing documents
```

```python
mycol4.insert_many(a) #inserting the resulted document in new collection
```

```python
for i in mycol4.find():
    print(i)
```
